Git is a very powerful tool when working with data. In group projects, it is likely that data can be lost in the seemingly endless versions of a final product or deleted entirely. Git is the perfect tool to maintain proper data management. 

Git can save information pertaining to who made changes, what changes were made, when were changes made, and why changes were made all through an iterative process.

### Before we Start

#### Common terms and phrases



**Repositories** - The collection of all your project's files. The repository also stores your history/versions

**Commits** - The snapshots (or versions) of files

**Branches** - Are different versions of the repository. The main branch encompesses what may be thought as the final project. New branches can be seen as areas for safe editing that could then add features to the main branch later

**Pull Requests** - Inform others of the changes you've made (pushed) to a certain branch. You can set it so pull requests must be reviewed first

### Cloning Repository & Creating Branch

Now that the repository is created, open your command prompt, activate `sia`, and navigate to where you would like to place your repository.

- `git clone https://github.com/YOUR-USERNAME/YOUR-REPOSITORY`

Navigate to your newly cloned repository
- `cd/YOUR-REPOSITORY`

The README.md file should be the only file within it.

Now we want to make our own branch within the repository. 

Check which branch you are now with:
- `git status`

It should say:
- `On branch main`

Create your own branch with:
- `git checkout -b YOUR-BRANCH-NAME`

Follow up with another
- `git status`

To switch back and forth between branches use:
- `git checkout NAME-OF-BRANCH`

### First File

For this tutorial, create an empty text file for your first commit.

In the repository input:
- `type nul > hello.txt`

While the hello.txt is in your folder, it is not a part of your repository yet. We need to transfer your file from working directory to the staging area. We want to tell Git that certain files will be a part of the next commit.

Input:
- `git add hello.txt`



We are now going to be commiting everything in the staging area. Make sure to add a suitable comment:
- `git commit -m "YOUR-COMMENT"`

### Push Your Files
Now we want to send all the files from your branch into the main repository.
Input:
- `git push`

If this results with an error, you may need to input:
- `git push --set-upstream origin YOUR-BRANCH-NAME`

```python
Image("images/Push.png")
```

### Pull Requests
Return to your browser to view the repository. You should notice at the top that your branch has a recent push.
- Near the top of the window select `Pull requests`
- Select `New pull request`
- Select your branch
- Review the files that will be added to your main branch and select `Create pull request`
- Leave a comment on this request and select again `Create pull request`
- Then finally select `Merge pull request`

Return to your repository page and make sure you are on the main branch. Your newly added files should be located there
