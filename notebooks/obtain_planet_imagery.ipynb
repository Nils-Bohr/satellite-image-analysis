```python
import os
import planet
import json
import requests
from requests.auth import HTTPBasicAuth

PLANET_API_KEY = 'PLAK792d76a9c70b481ba9854bf3353cc649'
item_type = "PSScene4Band"

geojson_geometry = {
  "type": "Point",
  "coordinates": [-78.378319, 38.799560],  
}

geometry_filter = {
  "type": "GeometryFilter",
  "field_name": "geometry",
  "config": geojson_geometry
}

date_range_filter = {
  "type": "DateRangeFilter",
  "field_name": "acquired",
  "config": {
    "gte": "2022-08-01T00:00:00.000Z",
    "lte": "2022-08-25T00:00:00.000Z"
  }
}

cloud_cover_filter = {
  "type": "RangeFilter",
  "field_name": "cloud_cover",
  "config": {
    "lte": 0.1
  }
}

combined_filter = {
  "type": "AndFilter",
  "config": [geometry_filter, date_range_filter, cloud_cover_filter]
}

search_request = {
  "item_types": [item_type], 
  "filter": combined_filter
}

search_result = requests.post(
    'https://api.planet.com/data/v1/quick-search',
    auth=HTTPBasicAuth(PLANET_API_KEY, ''),
    json=search_request)

results = search_result.json()

image_ids = [feature['id'] for feature in results['features']]
image_ids
id0 = image_ids[0]

id0_url = 'https://api.planet.com/data/v1/item-types/{}/items/{}/assets'.format(item_type, id0)

result = requests.get(
    id0_url,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )

links = result.json()["analytic"]["_links"]

#print(links)
# print(result.json())
self_link = links["_self"]
activation_link = links["activate"]

activate_result = requests.get(
    activation_link,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )
activate_result
activation_status_result = requests.get(
    self_link,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )
    
print(activation_status_result.json()["status"])
print(activation_status_result.json())

# Image can be downloaded by making a GET with your Planet API key, from here:
download_link = activation_status_result.json()["location"]
print(download_link)
```
