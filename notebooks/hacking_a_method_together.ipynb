# Hacking a method together

In this tutorial we're going to explore how to hack a method together. 

Whereas the lit review approach can be iterative, the method will almost certainly require multiple iterations/revisions. 

Why?

Essentially, because you will try an initial method and may find it does not achieve sensible and/or unbiased results. Moreover, you also may find that you need to undertake more parameterization or validation, once you have your first set of results, which requires improving the method. 

Therefore, you can see any methodology section as an evolving document. It is advisable to use your literature review to explore potential options, and then put pen to paper (metaphorically), to sketch out a potential method. At this stage, bullet points are fine for a first step. 

It is expected that a method will broadly have the following components:

- Description/visualization of the key method steps.
- Data sources to be used (including proper cited references to the sources).
- A detailed explanation of any pre-processing of your data.
- Consideration of any data validation steps (if required).


## Step 1 Method steps

The first thing you will want to do in your method is describe the general approach being taken. One highly effective way to do this is by using a box diagram. Why this is a good idea, will now be explained.

Scientific analysis needs to be reproducible. This means that any scientific paper published by a researcher should allow another separate researcher to reproduce the findings, based on following the published method. The beauty of a box diagram is that it provides insight on all input and output data, along with any processing used in the analysis. It is then possible to clearly link the method structure to the stages in the diagram. 

Let's take a look at the approach in a recent paper, entitled [predicting cell phone adoption metrics using machine learning and satellite imagery](https://www.sciencedirect.com/science/article/pii/S0736585321000617), published in Telematics and Informatics (impact factor: 9.1).

## Data sources  

You want to provide a full, comprehensive overview of the data sources you intend to use. This might include:

- Primary data, e.g., from Planet or Sentinel-2.
- Secondary data, e.g., other environmental data on rain fall, weather patterns etc.

Not only should you provide a complete citation to your source, but you should also describe the data. For example, below is a table from the aforementioned  Telematics and Informatics paper ([here](https://www.sciencedirect.com/science/article/pii/S0736585321000617)).

You will see that the survey training data reported is broken down by country, variable description, data file name, column metric, the number of households surveyed and the number of point clusters present. This is a strong amount of information to help the analysis be reproduced, and to help the reader understand the key statistical dimensions of the used data (e.g., sample size).

## Data preprocessing

The likelihood is that you will need to carry out various preprocessing steps for your data. 

For example, satellite image data which is obtained from a server may well be in the form of a very large tile, and therefore you may decide to clip each asset to an Area of Interest (AoI).

Additionally, any synthetic metrics produced which utilize multiband imagery will require these bands to be reordered or processed in a specific way. If this is the case, my suggestion is that you make sure you add in an equation for the processing carried out. 

Why? Well some people prefer graphical representation (e.g., the box diagram presented above) and some people prefer a mathematical description. You are on safe ground if you provide both. However, if you skip the equation, some scientists will not like this, and may criticise you for not being rigorous enough.

## Reflecting on this information

It is highly advisable to structure your method around the key steps required to answer your research question. If you are stuck on how to do this, then examine how others have carried out this process. 

For example, within the [paper](https://www.sciencedirect.com/science/article/pii/S0736585321000617) highlighted, the method is broken into six sections, with many of those key sections highlighted in the box diagram provided. Such an approach will enable the reader to easily follow along with the method, both in the text of the paper/report, and vis the graphical representation. 

Any case study information you need to provide about the location you are focusing is also best off going in your method at the end. It is recommended to keep your work as broad and general as possible in the introduction and literature review, only focus into a case study area later in the paper as an example of the phenomenon of interest. 

This is because papers that are marketed as case studies have been found statistically to receive fewer citations, as other researchers may find they are less relevant to their area of interest. 

Additionally, scientific analysis is about being able to say something systematic and statistically representative about a phenomenon, and case study research can make these aims challenging at times. Do try remember this.
